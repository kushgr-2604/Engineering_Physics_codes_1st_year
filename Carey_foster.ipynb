```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

R1=float(input("enter the value of Resistance R1:"))
balancing_length1=float(input("enter the balancing length of left Side:"))
balancing_length2=float(input("enter the balancing length of Right Side:"))
X=balancing_length2-balancing_length1
rho = R1/X
data=[[R1, balancing_length1, balancing_length2, X, rho]]
df=pd.DataFrame(data, columns=['Resistane ', 'position of balance point l1',
                 'position of balance point l2',
                 'l1-l2',
                 'Resistance per unit length'], index=[0])
print(df)
```
