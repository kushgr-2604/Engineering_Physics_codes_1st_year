**Aim:**To determine the high resistance by Leakage Method.

```python
import math

import pandas as pd
import numpy as np
C=1*10**(-6)
t=[]
volt1=[]
volt2=[]
R=[]
for i in range(4):
  a=int(input("Enter time in sec :"))
  t.append(a)
  b=float(input("Enter first voltage :"))
  volt1.append(b)
  c=float(input("Enter second voltage :"))
  volt2.append(c)
for i in range(4):
  d=1/(2.303*(math.log(volt1[i]/volt2[i]))*(C/t[i]))
  R.append(d)
df=pd.DataFrame({"Time":t,"First Voltage":volt1,"Second Voltage":volt2,"Resistance":R})
print(df)
```

**Result :**
Value of unknow resistance =4.902491e+06 ohm

**Precation :**
1. connection must be tight
2.Capacitor charging time should be minimum
