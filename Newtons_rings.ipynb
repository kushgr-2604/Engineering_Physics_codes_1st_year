**Theory:**
If a plano-convex lens of large radius of curvature is placed with its convex surface in contact with a plane glass plate, an air film of gradually increasing thickness is formed between them. If monochromatic light is allowed to fall normally and viewed, as shown in Fig.1, then alternate dark and bright circular fringes or rings are observed. These rings are known as Newtonâs rings. These fringes are formed due to the interference between the waves reflected from the top and bottom surface of the air film formed between the plate and the lens.
By measuring the diameter of the Newtonâs rings, the wavelength of the light can be calculated.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
lhs_reading=[]
rhs_reading=[]
for l in range(10,0,-1):
  ms_lhs=int(input("Enter the vale of LHS main scale:"))
  vs_lhs=int(input("Enter the vale of LHS vernier scale:"))
  total_reading_lhs=ms_lhs+(0.01*vs_lhs)
  lhs_reading.append(total_reading_lhs)


for r in range(10,0,-1):
  ms_rhs=int(input("Enter the vale of RHS main scale:"))
  vs_rhs=int(input("Enter the vale of RHS vernier scale:"))
  total_reading_rhs=ms_rhs+(0.01*vs_rhs)
  rhs_reading.append(total_reading_rhs)

D=[]
for i in range(len(lhs_reading)):
  diff= lhs_reading[i]-rhs_reading[i]
  D.append(diff)
D_sq=[]
for i in range(len(D)):
  D_sq.append(D[i]**2)
data_frame=pd.DataFrame({'Left hand Side(a) ': lhs_reading, 'Right hand Side(b) ': rhs_reading, 'Difference(a-b) ': D,'Diameter Square(a-b)^2':D_sq})
print(data_frame)
main_reading= []
for i in range(5):
  difference =D_sq[5 - i] - D_sq[9 - i]
  main_reading.append(difference)
print(main_reading)
```

```python
for i in main_reading:
  mean= np.mean(main_reading)
print(mean)
```
