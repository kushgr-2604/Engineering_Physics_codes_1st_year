**THEORY:**

 The electrical current passes through a meter filament of an incandescent lamp and the emission results into continuous spectrum, a characteristic of filament temperature. Stefan- Boltzmann relation (1) relates such a radiant energy of radiator by following relation.
E= σ AɛT ------------------------- (1)

A Surface area of filament
ɛ= Emissivity
ɛ= 1 (for black body)   
ɛ < 1 (for other bodies)
For a black body the radiant energy per unit area per second is given by
E = σT^4

However the power P(watt) in electrical filament circuit is the product of voltage drop V1 and current It gives loss of electrical energy per second. Hence on presumption that whole of P in converted to E, we can write modified form of relation (1) in term of P in the following order:
P=σ' T^4 -----------------------(2)

where σ' is a constant related to σ through the area of cross-section of the filament and radiation losses also note that the filament is not a black body.
In order to verify Stefan's law in term of equation (2), we take logarithm of this equation

```python
import pandas as pd
from logging import log
import numpy as np
import matplotlib.pyplot as plt
voltage= []
current= []
resistace= []
power= []
log_Power= []

n = int(input("Enter the number of readings to be inserted: "))
for i in range(n):
  v= float(input("enter the voltage :"))
  voltage.append(v)
  I=float(input("enter the current :"))
  current.append(I)
  Rt=(v/I)
  resistace.append(Rt)
  P=(v*I)
  power.append(P)
  log_Power.append(np.log10(P))

d={"Voltage": pd.Series(voltage),
   "Current":pd.Series(current),
   "Resistance" :pd.Series(resistace),
   "Power":pd.Series(power),
   "Power in Log":pd.Series(log_Power)
    }
df =  pd.DataFrame(d)
print(df)
```
