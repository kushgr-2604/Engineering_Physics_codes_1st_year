```python
from logging import log
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
n= int(input("Enter the nubers of reading taken :   "))
temp_in_C=[]

for x in range(n):
    value = float(input(f"Enter Temprautre in degree celsius {x+1}: "))
    temp_in_C.append(value)
print("Temprature in degree Celius")
print(*temp_in_C,sep="\n")

temp_in_K=[]
for x in temp_in_C:
    temp_in_K.append((273.15+x))
print("Temprature in Kelvin")
print(*temp_in_K,sep="\n")

inverse_temp_in_K=[]
for x in temp_in_K:
    inverse_temp_in_K.append(1000/x)
print("Inverse Temprature in Kelvin")
print(*inverse_temp_in_K,sep="\n")

current_in_microamp=[]
for y in range(n):
    value1 = float(input(f"Enter Current in microampere {y+1}: "))
    current_in_microamp.append(value1)
print("current in MicroAmpere")
print(*current_in_microamp,sep="\n")

log_current_in_microamp=[]
for x in current_in_microamp:
    log_current_in_microamp.append(np.log10(x))
print("Log of current in Micro Ampere")
print(*log_current_in_microamp,sep="\n")
d=pd.DataFrame({"Temperature in celcius": pd.Series(temp_in_C),
   "Temperature in Kelvin":pd.Series(temp_in_K),
   "Inverse Temprature":pd.Series(inverse_temp_in_K),
   "Current":pd.Series(current_in_microamp),
  })
print(d)
plt.plot(inverse_temp_in_K,log_current_in_microamp,marker="*",scaley=True,scalex=True)
plt.grid(log )
plt.xlabel('inverse_temp_in_K')
plt.ylabel('log_current_in_microamp')
plt.title("Inverse Temprature vs Log of Current")
plt.show()
```
